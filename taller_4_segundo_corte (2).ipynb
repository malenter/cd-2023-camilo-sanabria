```python
df['Sex'] = df['Sex'].replace({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].replace({'S': 0, 'C': 1,'Q':2})
```

```python
x =  df.iloc[:,[2,4,5,6,10]]
y =  df.Survived
logit_model=sm.Logit(y,x)
result=logit_model.fit()
print(result.summary())
```

```python
import itertools
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


X = df.iloc[:,[2,4,5,6,10]] 
y = df.iloc[:,df.Survived]  

combinations = list(itertools.combinations(X.columns, 5))

best_r2 = -np.inf
for combo in combinations:
    temp_X = X[list(combo)]
    regressor = LinearRegression()
    regressor.fit(temp_X, y)
    y_pred = regressor.predict(temp_X)
    r2 = r2_score(y, y_pred)
    if r2 > best_r2:
        best_r2 = r2
        best_vars = list(combo)


print("Mejor combinación de variables: ", best_vars)
print("Valor de R2 correspondiente: ", best_r2)
```

```python
from sklearn.model_selection import train_test_split

#Predicciones con scikit-learn
train, test = train_test_split(df, test_size = 0.20)
print("Ejemplos usados para entrenar: ", len(train))
print("Ejemplos usados para test: ", len(test))
```

```python
#Se definen las variables
x_train = train.iloc[:,[4,5,6,7]]
y_train = train.Survived
x_test = test.iloc[:,[4,5,6,7]]
y_test = test.Survived
```

```python
logit_model=sm.Logit(y_train, x_train)
result=logit_model.fit()
print(result.summary())
```

```python
yhat = result.predict(x_test)
prediction = list(map(round, yhat))
  
# Comparar l
print('Actual values:', list(y_test))
print('Predictions :', prediction)
y_test_list = list(y_test)
prediction_list = list(prediction)

```

```python
df.info()
```

```python
tp = 0
tn = 0
fp = 0
fn = 0

for i in range(len(y_test_list)):
    if y_test_list[i] == 1 and prediction_list[i] == 1:
        tp += 1
    elif y_test_list[i] == 0 and prediction_list[i] == 0:
        tn += 1
    elif y_test_list[i] == 0 and prediction_list[i] == 1:
        fp += 1
    elif y_test_list[i] == 1 and prediction_list[i] == 0:
        fn += 1

print("True positives:", tp)
print("True negatives:", tn)
print("False positives:", fp)
print("False negatives:", fn)
```

```python

presicion =tp/(fp+tp)
recall = tp/(fn+tp)

print("Precisión: ", presicion)
print("Recall: ", recall)
```

```python
import seaborn as sb
import os
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
predictores = df[['Pclass','Sex','Age','SibSp','Parch','Fare']]
objetivo = df.Survived
```

```python
pred_train, pred_test, res_train, res_test = train_test_split(predictores, objetivo, test_size=0.2)
```

```python
SVM_Model = SVC(gamma='auto', kernel='rbf')
SVM_Model.fit(pred_train, res_train)
```

```python
print (f'Precisión: {SVM_Model.score(pred_test, res_test):.4f}')
```

```python
from sklearn.metrics import recall_score

y_pred = SVM_Model.predict(pred_test)
recall = recall_score(res_test, y_pred, average='weighted')
print(f'Recall: {recall:.4f}')
```
