```python
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/adiacla/bigdata/master/titanic.csv")
df
```

```python
df.isna().sum()
```

```python
df = df.drop('Cabin', axis = 1)
df
```

```python
df = df.dropna()
df
```

```python
df.info()
```

```python

categorias = df['SibSp'].unique()
print(categorias)
```

```python
categorias = df['Embarked'].unique()
print(categorias)
```

```python
df['Sex'] = df['Sex'].replace({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].replace({'S': 0, 'C': 1,'Q':2})
```

```python
df.info()
```

```python
x =  df.iloc[:,[2,4,5,6,10]]
y =  df.Survived
logit_model=sm.Logit(y,x)
result=logit_model.fit()
print(result.summary())
```

```python
import itertools
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


X = df.iloc[:,[2,4,5,6,10]] 
y = df.iloc[:,df.Survived]  

combinations = list(itertools.combinations(X.columns, 5))

best_r2 = -np.inf
for combo in combinations:
    temp_X = X[list(combo)]
    regressor = LinearRegression()
    regressor.fit(temp_X, y)
    y_pred = regressor.predict(temp_X)
    r2 = r2_score(y, y_pred)
    if r2 > best_r2:
        best_r2 = r2
        best_vars = list(combo)


print("Mejor combinación de variables: ", best_vars)
print("Valor de R2 correspondiente: ", best_r2)
```
